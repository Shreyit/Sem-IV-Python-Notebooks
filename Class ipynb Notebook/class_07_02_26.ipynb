## **Dictionaries**
- uses {}
- stores key-value pairs
- mutable
- keys must be unique
- used in APIs & databases

```python
d = {}
type(d)
```

```python
d = {'roll.no':[1,2,3,4,5], 'marks':[66,77,56,87,97]}
d
```

```python
d['roll.no'] = 6
d.update( {'gpa': 7})
d
```

```python
d['roll.no'] = [1,2,3,4,5,6]
d
```

```python
print(d.keys())
print(d.values())
print(d.items())
print(d.get('marks'))
```

```python
x = 'I like Java'
x.replace('Java', 'Python')
print(x)
```

```python
y = ['c++', 'Python', 'Java', 'c#']
y
```

```python
y.extend(['Ruby', 'JavaScript'])
print(y)
```
