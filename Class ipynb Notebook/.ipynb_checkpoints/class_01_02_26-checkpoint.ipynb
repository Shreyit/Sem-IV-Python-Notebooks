```python
import math
import random

print(math.sqrt(25))
print(math.factorial(5))
print(math.ceil(2.5))        ## round to highest no. i.e. 2.5 or 2.3 will be 3
print(math.floor(2.5))       ## round to lowest no. i.e. 2.5 or 2.3 will be 2

print(random.randint(1,10))
print(random.random())
```

## **String**
- Immutable
- Indexing accdesses char by position
- Slicing extracts substring
- String methods help in formatting and searching

```python
a = 'First_String'
len(a)                  ## Full Indexing Length
```

```python
a[5]                    ## Indexing
```

```python
a[7]
```

```python
a[-4]                   ## Backward Indexing, starts from - [-1]
```

```python
s = "Python Learning"
s[0:8]                  ## Slicing
```

```python
s[:4]                   ## Indexing 0-3
```

```python
s[4:]                   ## Indexing 4-last
```

```python
s[4:-1]                 ## Indexing 4-(last-1)
```

```python
s.upper()               ## Uppercase
s.lower()               ## Lowercase
s.title()               ## Capitalizes first letter of each word
s.find("Learn")         ## will give the index from where it is starting
s.replace("Learning","Practice")
```

```python
s = "python_learning_is_good"
s.split("_")            ## will give - ['python', 'learning', 'is', 'good']
```

```python
cust_l = "00005"
cust_l.strip("0000")
```
