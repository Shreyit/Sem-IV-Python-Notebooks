## **SETS**
- the elements inside must be immutable, set is mutable
- every element is unique (they remove duplicates automatically)
- usage: 1. unique datasets which needs to be compared | 2. data cleaning
- **&** - (intersection)
- **|** - (union)

```python
setA = set("aabbccddeeVVgg")
setB = set("zzaaXXnnccyy")
print("setA - ", setA)
print("setB - ", setB)
```

```python
print("intersection - ", setA & setB)     # Common value from boath the sets
print("union - ", setA | setB)            # Unique form both the sets 
print("difference - ", setA - setB)
print("difference - ", setB - setA)
```

```python
setA.add(10)  # shows set can be mutable
setA
```

```python
setA.remove('V')
setA
```

## **COMPOUND STATEMENTS**
- *if*
- *elif*
- *else*

```python
if False:
    print("if condition is correct")
else:
    print("if statement was wrong so it jumps to else condition")
```

```python
if False:
    print("if condition is correct")
elif False:
    print("elif condition is correct")
elif True:
    print("2nd elif condition is correct")
else:
    print("if statement was wrong so it jumps to else condition")
```

## **Question:**
## Write a Pyhton programme that takes a students marks as input and prints the grade according to the following rules:
#### If marks are 90 or above print A, if marks are from 89 to 75 then print B, if 74 to 60 then print C, rest fail.

```python
marks = int(input('Enter your marks to check grade: '))

if marks >= 90:
    print('Grade A')
elif marks >= 75:
    print('Grade B')
elif marks >= 60:
    print('Grade C')
else:
    print('Fail')
```
