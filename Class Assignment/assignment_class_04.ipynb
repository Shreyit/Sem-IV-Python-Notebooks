## DICTIONARY

```python
tiss = {'roll.no':[1,2,3,4,5], 'marks':[66,77,56,87,97]}
tiss
```

```python
# 1.Update record
# update() function allows us to change the dictionary by specifying the new value for the key.
tiss.update({'gpa': [7.0, 7.7, 8.0, 7.8, 8.4]})
tiss
```

```python
key_count = tiss.keys()
print ("Total no. of keys = ", len(key_count))
```

```python
# get() retrieves a value safely; returns None if the key is missing
marks_list = tiss.get('marks')
marks_list
```

```python
# pop() removes the specified key and returns its value
removed_gpa = tiss.pop('gpa')
removed_gpa
```

```python
# Item Function returns a view object containing the dictionary's key-value pairs
print(f"Items in tiss: {tiss.items()}")
```

```python
# Delete key: Use 'del' to remove a specific record
del tiss['roll.no']
tiss
```

```python
simple_scores = {'Shreyit': 85, 'Sherick': 70, 'Satvik': 95}

# Sorts the dictionary by its values (scores) in ascending order
sorted_scores = dict(sorted(simple_scores.items(), key=lambda item: item[1]))
print(f"Sorted scores: {sorted_scores}")
```

## LIST

```python
y = ['c++', 'Python', 'Java', 'c#']

# 7. Append
y.append('Ruby') # Adds 'Ruby' to the end
y
```

```python
# 8. Insert
y.insert(1, 'SQL') # Adds 'SQL' at index 1
y
```

```python
# 9. Pop
last_lang = y.pop() # Removes 'Ruby'
y
```

```python
# 10. Remove
y.remove('c++') # Searches and removes 'c++'
y
```

```python
# 11. Extend
y.extend(['Kotlin', 'Go']) # Adds both to the end
y
```

```python
# 12-16. Statistical functions
nums = [66, 77, 56, 87, 97, 77]
print(len(nums))      # (Total items)
print(max(nums))      # (Highest value)
print(min(nums))      # (Lowest value)
print(nums.count(77)) # (Frequency of 77)
print(nums.index(87)) # (Position of 87)
```
